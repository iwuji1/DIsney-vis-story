```python
import pandas as pd 
import numpy as np

df = pd.read_csv("disney_movies_groups.csv", header=0, encoding='utf-8')
print (df.head())
```

```python
df[['inflation_adjusted_gross', 'time']].groupby('time').mean().sort_values(by=['inflation_adjusted_gross'],ascending=False)
```

```python
df[['inflation_adjusted_gross', 'genre']].groupby('genre').mean().sort_values(by=['inflation_adjusted_gross'], ascending=False)
```

```python
df[['inflation_adjusted_gross', 'genre']].groupby('genre').max().sort_values(by=['inflation_adjusted_gross'], ascending=False)
```

```python
df[['movie_title','inflation_adjusted_gross','time','genre']].sort_values(by=['inflation_adjusted_gross'], ascending=False).head()
```
