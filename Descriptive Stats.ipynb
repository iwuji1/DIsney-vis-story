```python
import pandas as pd 
import numpy as np

df = pd.read_csv("disney_movies_groups.csv", header=0, encoding='utf-8')
print (df.head())
```

```python
df[['inflation_adjusted_gross', 'time']].groupby('time').count().sort_values(by=['inflation_adjusted_gross'],ascending=False)
```

```python
df[['inflation_adjusted_gross', 'genre']].groupby('genre').mean().sort_values(by=['inflation_adjusted_gross'], ascending=False)
```

```python
df[['inflation_adjusted_gross', 'genre']].groupby('genre').count().sort_values(by=['inflation_adjusted_gross'], ascending=False)
```

```python
df[['inflation_adjusted_gross', 'mpaa_rating']].groupby('mpaa_rating').mean()
```

```python
df[['inflation_adjusted_gross', 'mpaa_rating']].groupby('mpaa_rating').count()
```

```python
df[['inflation_adjusted_gross', 'genre']].groupby('genre').count()
```
